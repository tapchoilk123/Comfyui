```python
import os
import subprocess
import time
import signal
import sys


def run_command(command, step_description, critical=False):
    """
    Execute a shell command and print the step's description and output.
    Exit the script if critical command fails.
    """
    print(f"=== {step_description} ===")
    result = subprocess.run(command, shell=True, text=True, capture_output=True)
    if result.returncode == 0:
        print(f"--- Success: {step_description} ---\n{result.stdout}")
    else:
        print(f"--- Failed: {step_description} ---\n{result.stderr}")
        if critical:
            print(">>> Critical step failed. Exiting...")
            sys.exit(1)
    print("======================================\n")
    return result.returncode, result.stdout, result.stderr


def start_ollama_serve():
    """
    Start Ollama in serve mode and verify if it's running.
    """
    print(">>> Starting Ollama in serve mode...")
    process = subprocess.Popen("ollama serve", shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, preexec_fn=os.setsid)
    time.sleep(5)  # Allow some time for the server to initialize

    # Verify the service is running
    print(">>> Verifying Ollama API...")
    result = subprocess.run("curl -s 127.0.0.1:11434", shell=True, text=True, capture_output=True)
    if result.returncode == 0:
        print(">>> Ollama API is running!")
    else:
        print(">>> Ollama API is not accessible. Please check for issues.")
        stop_ollama(process)
        sys.exit(1)

    return process


def stop_ollama(process):
    """
    Stop the Ollama serve process by sending a termination signal.
    """
    print(">>> Stopping Ollama serve process...")
    os.killpg(os.getpgid(process.pid), signal.SIGTERM)
    print(">>> Ollama serve process stopped.")


def pull_and_test_model(model_name, test_prompt):
    """
    Pull a model from Ollama and test its functionality.
    """
    # Step 1: Pull the model
    print(f">>> Pulling model: {model_name}...")
    return_code, stdout, stderr = run_command(f"ollama pull {model_name}", f"Pulling {model_name} model", critical=True)

    # Step 2: Verify model
    print(f">>> Verifying model: {model_name}...")
    run_command("ollama list", f"Verifying available models after pulling {model_name}")

    # Step 3: Test the model API
    print(f">>> Testing {model_name} model API...")
    test_query = f'''curl -X POST -H "Content-Type: application/json" -d '{{"text":"{test_prompt}"}}' http://127.0.0.1:11434/api/v1/models/{model_name}'''
    return_code, stdout, stderr = run_command(test_query, f"Testing {model_name} model")
    if return_code == 0 and "404" not in stdout:
        print(f">>> {model_name} model test successful! Response:")
        print(stdout)
    else:
        print(f">>> {model_name} model test failed or returned 404. Response:")
        print(stderr)


# Step 1: Install Ollama
run_command("curl -s https://ollama.ai/install.sh | sh", "Installing Ollama", critical=True)

# Step 2: Start Ollama in Serve Mode
ollama_process = start_ollama_serve()

# Step 3: Pull and test LLAVA Model
pull_and_test_model("llava", "Describe the content of an image.")

# Step 4: Pull and test LLama 3.2 Model
pull_and_test_model("llama3.2", "What are the latest advancements in AI?")

# Step 5: Stop Ollama
stop_ollama(ollama_process)

# Step 6: Install pyngrok
run_command("pip install pyngrok", "Installing pyngrok")

# Step 7: Update ComfyUI
comfyui_path = "/workspace/ComfyUI"
if not os.path.exists(comfyui_path):
    print(f">>> Error: Path {comfyui_path} does not exist. Exiting.")
    sys.exit(1)

print("-= Updating ComfyUI =-")
os.chdir(comfyui_path)
run_command("git branch --set-upstream-to=origin/main main", "Setting tracking branch for ComfyUI")
run_command("git pull", "Updating ComfyUI repository")

# Step 8: Move to custom_nodes directory
custom_nodes_path = os.path.join(comfyui_path, "custom_nodes")
os.chdir(custom_nodes_path)

if not os.path.isdir("ComfyUI-Manager"):
    print("-= Initial setup of ComfyUI-Manager =-")
    run_command("git clone https://github.com/ltdrdata/ComfyUI-Manager.git", "Cloning ComfyUI-Manager")

# Step 9: Install dependencies for ComfyUI-Manager
os.chdir("ComfyUI-Manager")
run_command("pip install GitPython", "Installing GitPython")
run_command("python cm-cli.py restore-dependencies", "Restoring dependencies for ComfyUI-Manager")

# Step 10: Return to ComfyUI root directory and list files
os.chdir(comfyui_path)
run_command("ls", "Listing files in ComfyUI root directory")

print(">>> Setup process completed successfully!")
```

```python
import subprocess
import threading
import time
import socket
import os
from pyngrok import ngrok


def kill_process_on_port(port):
    """
    Tự động tìm và dừng tiến trình đang sử dụng cổng.
    """
    try:
        result = subprocess.run(
            ["lsof", "-ti", f":{port}"],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )
        pid = result.stdout.strip()
        if pid:  # Nếu tìm thấy PID
            print(f"Found process {pid} using port {port}. Terminating it...")
            os.system(f"kill -9 {pid}")
        else:
            print(f"No process found using port {port}.")
    except Exception as e:
        print(f"Error while killing process on port {port}: {e}")


def monitor_server_with_logging(port, comfyui_process, max_wait=120):
    """
    Kiểm tra khi nào server chạy và in log trong quá trình kiểm tra.
    """
    print(f"Checking if server is running on port {port}...")
    waited = 0

    def log_comfyui():
        """
        In log từ stdout và stderr của ComfyUI ra console, bỏ qua lỗi ALSA.
        """
        while comfyui_process.poll() is None:
            line = comfyui_process.stdout.readline()
            if line.strip():
                print(f"[ComfyUI] {line.strip()}")
            error = comfyui_process.stderr.readline()
            if error.strip() and "ALSA" not in error:  # Bỏ qua lỗi ALSA
                print(f"[ComfyUI][ERROR] {error.strip()}")

    # Chạy in log trong thread riêng
    threading.Thread(target=log_comfyui, daemon=True).start()

    # Kiểm tra server
    while waited < max_wait:
        time.sleep(1)
        waited += 1
        try:
            sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
            result = sock.connect_ex(('127.0.0.1', port))
            if result == 0:  # Server sẵn sàng
                print(f"Server is ready on port {port}.")
                return True
        finally:
            sock.close()

    print(f"Server not ready on port {port} after {max_wait} seconds.")
    return False


def start_ngrok_tunnel(port):
    """
    Khởi động Ngrok tunnel với authtoken.
    """
    ngrok.set_auth_token("2pcsJqgx0uuY7OWIDOAeos9LMbB_2kQzqjVPjMQso9KJrL2uv")
    print("\nServer is ready. Launching Ngrok tunnel...\n")
    public_url = ngrok.connect(port)
    print("Ngrok tunnel is ready. Access your service at:", public_url)
    return public_url


def start_ollama():
    """
    Khởi chạy Ollama server.
    """
    print("\nStarting Ollama serve...\n")
    process = subprocess.Popen(
        ["ollama", "serve"],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    for line in process.stdout:
        print(line.strip())


def start_comfyui():
    """
    Khởi chạy ComfyUI server và trả về process.
    """
    print("Starting ComfyUI server...")
    process = subprocess.Popen(
        ["python3", "main.py"],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )
    return process


# --- Main logic ---
kill_process_on_port(8188)  # Port cho ComfyUI server
kill_process_on_port(11434)  # Port mặc định của Ollama serve

# Thay đổi thư mục làm việc tới ComfyUI
comfyui_path = '/workspace/ComfyUI'
os.chdir(comfyui_path)
print("Current working directory:", os.getcwd())

# Khởi động ComfyUI và in log
comfyui_process = start_comfyui()

# Kiểm tra server và in log trong quá trình kiểm tra
if monitor_server_with_logging(8188, comfyui_process):
    print("ComfyUI server started successfully.")

    # Khởi chạy Ngrok khi server sẵn sàng
    url = start_ngrok_tunnel(8188)
    if url:
        print(f"Access your service at: {url}")

        # Sau khi Ngrok hoạt động, khởi chạy Ollama
        start_ollama()
    else:
        print("Failed to create Ngrok tunnel.")
else:
    print("ComfyUI server failed to start. Please check the logs.")
```
